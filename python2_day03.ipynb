0. 복습
1) 크롤링 : 웹 페이지를 제어하는 기술

- 웹 크롤러 : 웹 사이트에 있는 수많은 정보 가운데 우리가 원하는 정보를 수집하는 프로그램
	=> 웹 사이트의 정보를 수집하기 위해 작성한 코드를 웹 크롤러,
	     이를 이용해서 데이터를 수집하는 행위를 웹 크롤링
- 웹 스크래핑 : 웹 사이트에서 원하는 정보만을 일부 추출 한다는 의미

2) 웹 크롤링 과정
(1) 정보를 얻고자하는 웹 사이트에 접속해서 웹 페이지를 확인
(2) 키보드에서 개발자모드(F12)를 눌러 내가 원하는 정보의 위치를 확인하고 분석
(3) 파이썬 코드를 작성해서 접속한 웹 페이지의  html 코드를 불러옴
(4) 불러온 데이터에서 원하는 정보를 가공한 후 추출

3) HTTP(HyperText Transfer Protocol) : 웹 페이지에서 데이터를 전달할 때 사용하는 프로토콜

4) HTML(HyperText Markup Language) : 하이퍼텍스트와 마크업언어를 합친 언어
		
5) 크롤링 시 주의사항
 무분별한 웹 크롤링은  웹서버에 과부하 문제를 유발할 수 있으므로 크롤링 하기 전에 항상 robots.txt 파일 확인할것!!!

```python
import requests
from bs4 import BeautifulSoup as bs

url = 'https://www.melon.com/'
response = requests.get(url)

print(response.status_code) #4xx 코드 : 사용자 잘못
print(response.text) #상태코드가 200이 아니므로 출력안됨 -> 헤더 추가 작성 필요함
```

```python
import requests
from bs4 import BeautifulSoup as bs

url = 'https://www.melon.com/'
header = {"User-Agent" : "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/105.0.0.0 Safari/537.36"}
response = requests.get(url, headers = header)

print(response.status_code) #406 : 사용자 잘못
print(response.text)
```

```python
import requests
from bs4 import BeautifulSoup as bs

url = 'https://vibe.naver.com/chart/total'
html = requests.get(url)
soup = bs(html.content, 'html.parser')

print(soup.text)
```
